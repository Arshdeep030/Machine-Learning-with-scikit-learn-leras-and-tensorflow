```python
df["rooms_per_household"] = df["total_rooms"]/df["households"]
df["bedrooms_per_room"] = df["total_bedrooms"]/df["total_rooms"]
df["population_per_household"]=df["population"]/df["households"]
```

```python

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.base import BaseEstimator, TransformerMixin

#  Custom transformer to add new features
class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room=True):
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        rooms_per_household = X[:, 0] / X[:, 3]
        population_per_household = X[:, 2] / X[:, 3]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, 1] / X[:, 0]
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]

#  Separate columns
num_cols = ["total_rooms", "total_bedrooms", "population", "households", "median_income"]
cat_cols = ["ocean_proximity"]

#  Numerical pipeline
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('attribs_adder', CombinedAttributesAdder()),
    ('scaler', StandardScaler())
])

#  Full preprocessing pipeline
full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_cols),
    ("cat", OneHotEncoder(), cat_cols)
])

#  Apply to the housing dataset
housing= full_pipeline.fit_transform(df)

#  Done: housing_prepared is ready for model training
print(housing.shape)
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

num_attribs = list(num_cols)
cat_attribs = ["ocean_proximity"]

class OldDataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names].values

old_num_pipeline = Pipeline([
    ('selector', OldDataFrameSelector(num_attribs)),         # Select numerical columns
    ('imputer', SimpleImputer(strategy="median")),           # Fill missing values
    ('attribs_adder', CombinedAttributesAdder()),            # Add new features
    ('std_scaler', StandardScaler()),                        # Normalize (mean=0, std=1)
])
```

```python
old_cat_pipeline = Pipeline([
    ('selector', OldDataFrameSelector(cat_attribs)),     # Select categorical column(s)
    ('cat_encoder', OneHotEncoder(sparse_output=False)),        # One-hot encode the category
])
```

```python
from sklearn.pipeline import FeatureUnion

old_full_pipeline = FeatureUnion(transformer_list=[
    ("num_pipeline", old_num_pipeline),
    ("cat_pipeline", old_cat_pipeline),
])
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
import numpy as np

class TopFeatureSelector(BaseEstimator, TransformerMixin):
    def __init__(self, feature_importances, k):
        self.feature_importances = feature_importances
        self.k = k

    def fit(self, X, y=None):
        self.feature_indices_ = np.argsort(self.feature_importances)[-self.k:]
        return self

    def transform(self, X):
        return X[:, self.feature_indices_]
```

```python
from sklearn.ensemble import RandomForestRegressor

# Fit the full pipeline (without feature selector yet)
housing_prepared = full_pipeline.fit_transform(X_train)

# Train a model to get importances
forest = RandomForestRegressor(random_state=42)
forest.fit(housing_prepared, y_train)

# Get importances
feature_importances = forest.feature_importances_
```

```python
from sklearn.pipeline import Pipeline

k = 10  # Choose top k features
top_k_feature_selector = TopFeatureSelector(feature_importances, k)

final_pipeline = Pipeline([
    ("preparation", full_pipeline),
    ("feature_selection", top_k_feature_selector),
    ("model", RandomForestRegressor(random_state=42))
])

final_pipeline.fit(X_train, y_train)

# Evaluate on test data
predictions = final_pipeline.predict(X_test)
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)

print(f"Test RMSE after feature selection: {rmse:.2f}")
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
import numpy as np

class TopFeatureSelector(BaseEstimator, TransformerMixin):
    def __init__(self, feature_importances, k):
        self.feature_importances = feature_importances
        self.k = k

    def fit(self, X, y=None):
        self.feature_indices_ = np.argsort(self.feature_importances)[-self.k:]
        return self

    def transform(self, X):
        return X[:, self.feature_indices_]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

k = 10  # number of top features to select

full_model_pipeline = Pipeline([
    ("preprocessing", full_pipeline),
    ("feature_selection", TopFeatureSelector(feature_importances, k)),  # Optional
    ("model", RandomForestRegressor(random_state=42))
])
```

```python
# Train
full_model_pipeline.fit(X_train, y_train)

# Predict
final_predictions = full_model_pipeline.predict(X_test)

# Evaluate
mse = mean_squared_error(y_test, final_predictions)
rmse = np.sqrt(mse)
print(f"Final RMSE: {rmse:.2f}")
```
